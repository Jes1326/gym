```python
df = df[['Age', 'Gender', 'Weight (kg)', 'Height (m)', 'Max_BPM', 'Avg_BPM',
       'Resting_BPM', 'Session_Duration (hours)', 'Calories_Burned',
       'Workout_Type', 'Fat_Percentage',
       'Workout_Frequency (days/week)', 'Experience_Level', 'BMI']].copy()
```

# Adding new column 
- based on ages - Teen, Adult, Senior

```python
df.loc[df['Age']<=20, 'Category'] = 'Teen'
df.loc[(df['Age']>20) & (df['Age']<=40), 'Category'] = 'Adult'
df.loc[df['Age']>40, 'Category'] = 'Senior'
```

- BMI Categories
-  Underweight: BMI < 18.5 ,Normal weight: 18.5 – 24.9, Overweight: 25 – 29.9, Obese: BMI ≥ 30

```python
df.loc[df['BMI']<=18.5, "Bmi_Category"] = "Underweight"
df.loc[(df['BMI']>18.5) & (df['BMI']<=24.9), "Bmi_Category"] = "Normalweight"
df.loc[(df['BMI']>25) & (df['BMI']<=29.9), "Bmi_Category"] = "Overweight"
df.loc[df['BMI']>30, "Bmi_Category"] = "Obese"
```

```python
df
```

Countplot of Different Category 

```python
plt.figure(figsize=(25,5))

plt.subplot(1,3,1)
sns.countplot(df,x='Category')
plt.title('Countplot of Different Category')
plt.ylabel('Count')

plt.subplot(1,3,2)
sns.lineplot(df,x=df['Workout_Frequency (days/week)'],\
             y=df['Workout_Frequency (days/week)'].value_counts(),\
             marker='o')
plt.title("Workout frequency Trend")
plt.ylabel("Count of members")
plt.xticks([3,4,5])

plt.subplot(1,3,3)
df['Bmi_Category'].value_counts().plot(kind='barh')
plt.ylabel('')
plt.title("DIstribution of BMI Category")

plt.show()
```

Distribution of members - Pie Chart

```python
plt.figure(figsize=(25,7))

plt.subplot(1,3,1)          
df['Category'].value_counts().plot(kind='pie',\
                                   autopct = lambda x: f'{x:.1f}% - {int(x * df.shape[0] / 100)}')
plt.title(label="Distribution of members")
plt.ylabel("")

plt.subplot(1,3,2)
df["Workout_Type"].value_counts().plot(kind='pie',\
                                       autopct=lambda p: f'{p:.1f}% - {int(p * sum(df["Workout_Type"].value_counts()) / 100)}')
plt.ylabel('')
plt.title("Distribution of Workout types")

plt.subplot(1,3,3)
df['Experience_Level'].value_counts().plot(kind='pie',\
                                           autopct= lambda p: f'{p:.1f} - {int(p * df.shape[0] / 100)}')
plt.title('Experience Level of Members(3,2,1)')
plt.ylabel('')

plt.show()
# lambda p: f'{p:.1f}% - {int(p * df.shape[0] / 100)}'
```

Weight Distribution - Histoplot

```python
plt.figure(figsize=(25,5))

bin_range = [40,50,60,70,80,90,100,110,120,130]
ax=sns.histplot(df,x=df['Weight (kg)'],bins=bin_range)
plt.title("Weight Distribution(in KG's)")
plt.xticks(bin_range)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2, height, int(height),\
            ha="center", va='bottom')


plt.show()
```

Category Vs BPM - Boxplot

```python
plt.figure(figsize=(25,5))

plt.subplot(1,3,1)
sns.boxplot(x='Category', y='Max_BPM', data=df, hue='Category')
plt.title("Max BPM Ditribution")

plt.subplot(1,3,2)
sns.boxplot(x='Category', y='Avg_BPM', data=df, hue='Category')
plt.title("Average BPM Ditribution")

plt.subplot(1,3,3)
sns.boxplot(x='Category', y='Resting_BPM', data=df, hue='Category')
plt.title("Resting BPM Ditribution")

plt.show()
```

Which Workout resulted in more calories burn?

```python
plt.figure(figsize=(25,5))

plt.subplot(1,2,1)
df.groupby("Workout_Type").agg("sum")['Calories_Burned'].sort_values().plot(kind='barh')
plt.xlabel('Calories Burned')
plt.title("Total Calories burned - Workout Type")

plt.subplot(1,2,2)
df.groupby("Workout_Type").apply(lambda x: x['Calories_Burned'].sum()/x['Session_Duration (hours)'].sum(), include_groups=False).plot(kind='barh')
plt.xlabel('Calories Burned/Session')
plt.title("Total Calories burned/Session")

plt.show()
```

How Calories Burned effects Session Duraion, Fat Percentage and

```python
plt.figure(figsize=(25,5))

plt.subplot(1,2,1)
sns.scatterplot(df,x=df['Calories_Burned'],\
                y=df['Session_Duration (hours)'],\
                hue=df['Workout_Type'])
plt.title("Correlation Analyis - Calories Burned Vs Session Duration")

plt.subplot(1,2,2)
sns.scatterplot(df,x=df['Calories_Burned'],\
                y=df['Fat_Percentage'],\
                hue=df['Category'])
plt.title("Correlation Analyis - Calories Burned Vs Fat Percentage")


plt.show()
```

What is the distribution of the BMI of member with respect to the Workout Frequency?

```python
plt.figure(figsize=(25,5))

plt.subplot(1,2,1)
sns.boxplot(df,y=df['BMI'],x=df['Workout_Frequency (days/week)'],\
            hue=df['Category'])
plt.title("BMI Vs Workout Frequency")

plt.subplot(1,2,2)
sns.boxplot(df,y=df['BMI'],x=df['Workout_Frequency (days/week)'],\
            hue=df['Gender'])
plt.title("BMI Vs Workout Frequency")

plt.show()
```

Average Calories Burned and Average Session Duration

```python
df.groupby('Workout_Type')\
    [['Calories_Burned','Session_Duration (hours)']].\
        mean().\
            round(2).\
            sort_values(by='Calories_Burned',ascending=False)
```

Average Session Duration

```python
float(df['Session_Duration (hours)'].mean().round(2))
```

Calculating Efficiency 

```python
df['Efficiency'] = df['Calories_Burned'] / df['Session_Duration (hours)'].round(2) #Defining Effeciency

plt.figure(figsize=(25,5))
plt.subplot(1,2,1)
sns.boxplot(df,x=df['Workout_Type'],y=df['Efficiency'],hue=df['Category'])

plt.subplot(1,2,2)
sns.boxplot(df,x=df['Category'],y=df['Efficiency'],hue=df['Gender'])

plt.show()
```
