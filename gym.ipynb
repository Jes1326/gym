```python
bins = [18,20,30,40,50,60]
labels = ['18-20','20-30','30-40','40-50','50-60']
age_bins = pd.cut(df['Age'], bins=bins, labels=labels, right=False, ordered=False).value_counts().sort_index()
```

```python
age_bins.plot(kind='bar')
```

```python
age_bins.plot(kind='barh',color='green',width=0.6)
plt.xlabel('Number of Members')
plt.show()
```

```python
import seaborn as sns
```

```python
df.plot(kind='scatter',x='Weight (kg)',y='Age')
```

```python
df.loc[df['Age']<=20, 'Category'] = 'Teen'
df.loc[(df['Age']>20) & (df['Age']<=40), 'Category'] = 'Adult'
df.loc[df['Age']>40, 'Category'] = 'Senior'
```

```python
df.head()
```

```python
sns.countplot(df,x='Category')
plt.show()
```

```python
cat_plot = sns.lineplot(x=df['Category'].value_counts().index,\
             y=df['Category'].value_counts().values,\
            marker= 'o',\
            color='red')
plt.show()
```

```python
cat_counts = df['Category'].value_counts().sort_index()
cat_plot = sns.lineplot(x=cat_counts.index,\
             y=cat_counts.values,\
            marker= 'o',\
            color='red')

for(x,y) in zip(cat_counts.index,cat_counts.values):
    plt.text(x,y,str(y),ha='left',va='bottom',fontsize=10)
plt.show()
```
