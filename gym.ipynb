```python
import seaborn as sns
```

```python
df.loc[df['Age']<=20, 'Category'] = 'Teen'
df.loc[(df['Age']>20) & (df['Age']<=40), 'Category'] = 'Adult'
df.loc[df['Age']>40, 'Category'] = 'Senior'
```

```python
df.head()
```

```python
sns.countplot(df,x='Category')
plt.show()
```

```python
cat_plot = sns.lineplot(x=df['Category'].value_counts().index,\
             y=df['Category'].value_counts().values,\
            marker= 'o',\
            color='red')
plt.show()
```

```python
cat_counts = df['Category'].value_counts().sort_index()
cat_plot = sns.lineplot(x=cat_counts.index,\
             y=cat_counts.values,\
            marker= 'o',\
            color='red')

for(x,y) in zip(cat_counts.index,cat_counts.values):
    plt.text(x,y,str(y),ha='left',va='bottom',fontsize=10)
plt.show()
```

```python
df['Category'].value_counts().plot(kind='pie',autopct = '%1.1f%%')
plt.title(label="Distribution of members")
plt.ylabel("")
# plt.legend(title="Categories")

plt.show()
```

```python
df['Category'].value_counts().plot(kind='pie',autopct = lambda x:f'{int(x * sum(df['Category'].value_counts())/100)}')
plt.title(label="Distribution of members")
plt.ylabel("")
# plt.legend(title="Categories")

plt.show()
```

```python
bin_range = [40,50,60,70,80,90,100,110,120,130]
ax=sns.histplot(df,x=df['Weight (kg)'],bins=bin_range)
plt.title("Weight Distribution(in KG's)")
plt.xticks(bin_range)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2, height, int(height),\
            ha="center", va='bottom')


plt.show()
```

```python
line = df[df['Gender']=='Male'][['Age','Max_BPM']].sort_values(by='Age')
sns.lineplot(line,x=line['Age'],y=line['Max_BPM'])
```

```python
teen_maxbpm_x=df[(df['Gender']=='Male') & (df['Category']=='Teen')][['Age','Max_BPM']].sort_values(by='Age')
sns.lineplot(teen_maxbpm_x,x=teen_maxbpm_x['Age'],y=teen_maxbpm_x['Max_BPM'],\
             errorbar=None,marker='s')
plt.title("Max_BPM of Teens (Male)")

teen_maxbpm_x=df[(df['Gender']=='Female') & (df['Category']=='Teen')][['Age','Max_BPM']].sort_values(by='Age')
sns.lineplot(teen_maxbpm_x,x=teen_maxbpm_x['Age'],y=teen_maxbpm_x['Max_BPM'],\
             errorbar=None,marker='s')
plt.title("Max_BPM of Teens (Female)")

Adult_maxbpm_x=df[(df['Gender']=='Male') & (df['Category']=='Adult')][['Age','Max_BPM']].sort_values(by='Age')
sns.lineplot(teen_maxbpm_x,x=teen_maxbpm_x['Age'],y=teen_maxbpm_x['Max_BPM'],\
             errorbar=None,marker='s')

plt.show()
```

```python
Adult_maxbpm_x=df[df['Category']=='Adult'][['Age','Max_BPM']]
sns.lineplot(teen_maxbpm_x,x=Adult_maxbpm_x['Age'],y=Adult_maxbpm_x['Max_BPM'],\
             errorbar=None,marker='s')
plt.xticks(np.arange(Adult_maxbpm_x['Age'].min(), Adult_maxbpm_x['Age'].max() + 1, 1))
plt.title("BPM in Adults(Workout)")

Adult_restingbpm_x=df[df['Category']=='Adult'][['Age','Resting_BPM']]
sns.lineplot(teen_maxbpm_x,x=Adult_restingbpm_x['Age'],y=Adult_restingbpm_x['Resting_BPM'],\
             errorbar=None,marker='s')
plt.xticks(np.arange(Adult_restingbpm_x['Age'].min(), Adult_restingbpm_x['Age'].max() + 1, 1))
plt.title("BPM in Adults(Rest)")

plt.show()
```
